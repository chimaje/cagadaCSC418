```python
import cv2
import os
import numpy as np 
import matplotlib.pyplot as plt
choice =0
def authenticate_user():
    email = input("Enter your email address: ")
    if not is_valid_email(email):
        print("Invalid email address.")
        return False
    
    age = int(input("Enter your age: "))
    if age < 18:
        print("You must be 18 years or older to access the museum.")
        return False
    
    print("Authentication successful!")
    return True

def is_valid_email(email):
    valid_email_domains = ["gmail.com", "yahoo.com", "hotmail.com"] 
    if "@" not in email or "." not in email:
        return False
    
    domain = email.split("@")[-1]
    return domain in valid_email_domains

def select_collection_category():
    print("Select the art collection category:")
    print("1. Traditional Art")
    print("2. Modern Art")
    print("3. Contemporary Art")
    choice = input("Enter the number corresponding to your choice: ")
    return int(choice)

def display_images(choice):
    if choice ==1:
        folder_path = "C:/Users/athen/Documents/Year 4/2nd Semester/CSC418/cagadaCSC418/Week3/TraditionalArt"
        category = " Traditional Art"
    elif choice ==  2 :
        folder_path = "C:/Users/athen/Documents/Year 4/2nd Semester/CSC418/cagadaCSC418/Week3/ModernArt"
        category = " Modern Art"
    elif choice == 3 :
        folder_path = "C:/Users/athen/Documents/Year 4/2nd Semester/CSC418/cagadaCSC418/Week3/ContemporaryArt"
        category = " Contemporary Art"
    else:
        print(choice)
        print("Invalid category.")
        return
    
    if not os.path.isdir(folder_path):
        print(f"Error: Folder '{folder_path}' not found.")
        return
    
    image_files = os.listdir(folder_path)
    for img_file in image_files:
        img_path = os.path.join(folder_path, img_file)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Unable to load image '{img_path}'.")
        else:
            while img is not None:
                cv2.imshow(category, img)
                cv2.waitKey(0)
                apply_transformation(img)
                
        cv2.destroyAllWindows()

def apply_transformation(image):
     while True:
        print("Select the transformation method:")
        print("1. Translate")
        print("2. Reflection")
        print("3. Rotated")
        print("4. Crop")
        print("5. Exit")
        
        choice = input("Enter the number corresponding to the transformation method: ")
        
        if choice == "1":
            # Translated image
            rows,cols = img.shape
            M=np.float32([[1,0,100],[0,1,50]])
            trans1 = cv.warpAffine(img,M,(cols,rows))
            cv2.imshow("Translated Image", trans1)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        elif choice == "2":
            # Reflection image
            rows,cols = image.shape
            M=np.float32([[1,0,0],[0,-1,rows],[0,0,1]])

            reflected_img = cv.warpPerspective(image,M,(int(cols),int(rows)))

            cv2.imshow("Reflected Image", reflected_img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        elif choice == "3":
            # Rotated image
            rows,cols = image.shape
            img_rotation = cv.warpAffine(image,cv.getRotationMatrix2D((cols/2,rows/2),30,0.6),(cols,rows))
            cv2.imshow("Rotated Image", img_rotation)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        elif choice == "4":
            # Cropped image
            cropped_img = image[450:700,200:500]
            cv2.imshow("Cropped Image", cropped_img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        elif choice == "5":
            print("Exiting transformation mode.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

def run_museum_app():
    if authenticate_user():
        category = select_collection_category()
        display_images(category)


run_museum_app()
```
