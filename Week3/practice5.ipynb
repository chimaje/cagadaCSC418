```python
#sheared image on X-axis
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

image = cv.imread('../image/sst.jpg',0)
rows,cols = image.shape
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)

M= np.float32([[1,0.5,0],[0,1,0],[0,0,1]])
sheared_img = cv.warpPerspective(image,M,(int(cols*1.5),int(rows*1.5)))
plt.subplot(1,2,2)
plt.title("Sheared Image on X-axis")
plt.imshow(sheared_img)

cv.imshow('img',sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
